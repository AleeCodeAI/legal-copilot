```python
from docling.document_converter import DocumentConverter
from pathlib import Path
```

```python
source = Path(__file__).parents[3] / "data" / "cleaned_2025_Landlord_Tenant_Guide.pdf"
# OR
source = "cleaned_2025_Landlord_Tenant_Guide.pdf" # if you used google colab and imported the pdf there
```

```python
converter = DocumentConverter()
```

```python
result = converter.convert(source)
```

```python
doc = result.document
```

```python
doc.save_as_json("data.json")
```
